```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Résultat 1er entrainement
df_results = pd.read_csv('/home/gaetansimplon/wildfire_detection/runs/detect/train/results.csv')
display(df_results.columns)
metrics_run1 = df_results[['metrics/precision(B)', 'metrics/recall(B)', 'metrics/mAP50(B)']]
display('Aperçu des métriques run 1 : ',metrics_run1.describe())
# display('Aperçu des 5 derniers epochs : ',df_results.tail())

df_box_loss = df_results[['epoch', 'train/box_loss', 'val/box_loss']]
losses_melt = df_box_loss.melt(id_vars="epoch", 
                          value_vars=["train/box_loss", "val/box_loss"], 
                          var_name="Type", value_name="Loss")

plt.figure(figsize=(10,6))
sns.lineplot(data=losses_melt,
             x='epoch',
             y='Loss',
             hue='Type')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.tight_layout()
plt.show()

```

# Modèle YOLOv9


## Run 4
### Paramètres : 

- Modèle : v9s (version small)
- epochs : 100
- imgsz : 512
- batch : 4

```python
# Résultats 4ème entrainement

df_results4 = pd.read_csv('/home/gaetansimplon/wildfire_detection/runs4/detect/train17/results.csv')
metrics_run4 = df_results4[['metrics/precision(B)', 'metrics/recall(B)', 'metrics/mAP50(B)']]
display('Aperçu des métriques run 4: ',metrics_run4.describe())
display(df_results4.columns)
```

## Visualisation des métriques

### Pertes (loss)

```python
## --- BOX LOSS --- ##
### erreur sur la localisation des boîtes (à quel point les prédictions s’alignent sur les objets). ###
df_box_loss = df_results4[['epoch', 'train/box_loss', 'val/box_loss']]
losses_melt = df_box_loss.melt(id_vars="epoch", 
                          value_vars=["train/box_loss", "val/box_loss"], 
                          var_name="Type", value_name="Loss")

plt.figure(figsize=(10,6))
sns.lineplot(data=losses_melt,
             x='epoch',
             y='Loss',
             hue='Type')
plt.title('Evolution du box_loss train vs val')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.tight_layout()
plt.show()
```

```python
## --- CLS LOSS --- ##
### erreur de classification (mauvais label prédit). ###
df_cls_loss = df_results4[['epoch', 'train/cls_loss', 'val/cls_loss']]
losses_melt = df_cls_loss.melt(id_vars="epoch", 
                          value_vars=["train/cls_loss", "val/cls_loss"], 
                          var_name="Type", value_name="Loss")

plt.figure(figsize=(10,6))
sns.lineplot(data=losses_melt,
             x='epoch',
             y='Loss',
             hue='Type')
plt.title('Evolution du classification_loss train vs val')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.tight_layout()
plt.show()
```
