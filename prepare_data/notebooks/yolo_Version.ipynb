### Visualisation du matrice des résultats des resultats du Yolo8

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from pathlib import Path

#  chemin vers ton dossier de résultats
run_dir = Path("/home/fadilatou/PROJETS/wildfire_detection/prepare_data/notebooks/train_yoloV/resultsV8")   

#  chemin vers la matrice de confusion
conf_matrix_path = run_dir / "/home/fadilatou/PROJETS/wildfire_detection/prepare_data/notebooks/train_yoloV/resultsV8/content/runs/detect/train3/confusion_matrix.png"

# Vérification
if conf_matrix_path.exists():
    img = mpimg.imread(conf_matrix_path)
    plt.figure(figsize=(9,9))
    plt.imshow(img)
    plt.axis("off")  # pas d’axes
    plt.title("Matrice de confusion YOLO")
    plt.show()
else:
    print(f"Fichier non trouvé : {conf_matrix_path}")

```
